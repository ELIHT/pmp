1 - Read the text file content

```python
with open('820058_vietnam_vs_saudi_arabia', 'r') as f:
    content = f.read()
    print(content)
```

2 - Split the content into a list of phrases. Print out the shortest length and the longest length of the
phrases.

```python
#run on terminal #$ pip install nltk
#run on terminal $ pip install pandas
#run on jupyter notebook/ Kernel/ Restart & Clear Output
```

```python
import nltk
```

```python
listphrases = nltk.sent_tokenize(content)
listphrases = [phrase.lower() for phrase in listphrases]
print('Longest phrase length: ', max([len(phrase) for phrase in listphrases]))
print('Shortest phrase length: ', min([len(phrase) for phrase in listphrases]))
```

3 - Use pandas, create a dataframe, the rows are for the phrases you got in the previous step, and
the cells are for words in phrases.

```python
import pandas as pd
import string
import numpy as np
```

```python
data = []
for phrase in listphrases:
    words = nltk.word_tokenize(phrase)
    no_row = {}
    no_col = 1
    for i in range(len(words)):
        if words[i] not in string.punctuation:
            no_row[f'[word{no_col}]'] = words[i]
            no_col += 1
    data.append(no_row)
df = pd.DataFrame(data)
df = df.replace(np.nan, '')
df
```

4 - What is the word the most frequent for each column.

```python
df = df.replace('', np.nan)
most_frequent = df.mode(dropna=True).head(1)
most_frequent
```

5 - Make a phrase that combines the most frequent words of the columns. Put it to the last row of
the dataframe.

```python
#Put to the last row of the dataframe
```

```python
df = df.replace(np.nan, '')
df = df.append(most_frequent, ignore_index=True)
df
```
